```python
novoindex=[] 
def cleanup(text):
    punctuation = '[!\ \n-.:?;,@]' # Note que os sinais [] são delimitadores de um conjunto.
    pattern = re.compile(punctuation)
    for frase in text:
         novoindex.append(re.sub(pattern, ' ', frase.lower()))
    return novoindex

# Por algum motivo o teu n funciona para a string da frase
```

```python
frase = 'era uma vez em hollywood é perfeito'
```

```python
lista=frase.split()
palavras=cleanup(lista)
palavras
```

```python
probabilidade_de_ser_boa=total_bom/total_valores
probabilidade_de_ser_boa
```

```python
probabilidade_de_ser_ruim=total_ruim/total_valores
probabilidade_de_ser_ruim
```

```python
probabilidade_de_ser_irrelevante=total_irrelevantes/total_valores
probabilidade_de_ser_irrelevante
```

## naive bayes

```python
p_bom=0
p_ruim=0
p_irrelevante=0
for i in palavras:
    if i not in tabela_freq2: 
        prob_bom[i]=0
    if i not in tabela_freq3a: 
        prob_ruim[i]=0
        tabela_freq3a[i]=1e-8
    if i not in tabela_freq4: 
        prob_irrelevante[i]=0
        tabela_freq4a[i]=1e-8
    p_bom+=(np.log(prob_bom[i]/x+(1/total_valores)))*probabilidade_de_ser_boa
    p_ruim+=(np.log(prob_ruim[i]/y+(1/total_valores)))*probabilidade_de_ser_ruim
    p_irrelevante+=(np.log(prob_irrelevante[i]/z+(1/total_valores)))*probabilidade_de_ser_irrelevante
    
print(p_bom)
print(p_ruim)
print(p_irrelevante)
```
