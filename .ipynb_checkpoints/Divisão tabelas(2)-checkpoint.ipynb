```python
probabilidade_de_ser_boa=total_bom/total_valores
probabilidade_de_ser_boa
```

## 1º jeito

```python
p=0
for i in palavras:
    if i not in tabela_freq3: 
        prob_bom[i]=0
    p+=(np.log(prob_bom[i]+1e-8))*probabilidade_de_ser_boa
print(p)
```

```python
p=0
for i in palavras:
    if i not in tabela_freq3: 
        prob_ruim[i]=0
    p+=(np.log(prob_ruim[i]+1e-8))*probabilidade_de_ser_ruim
print(p)
```

```python
p=0
for i in palavras:
    if i not in tabela_freq4: 
        prob_irrelevante=0
    p+=(np.log(prob_irrelevante[i]+1e-8))*probabilidade_de_ser_irrelevante
print(p)
```

## 2º jeito

```python
p=0
for i in palavras:
    if i not in tabela_freq3: 
        prob_ruim[i]=0
    p+=np.exp(np.log(prob_ruim[i]+1e-8)+np.log(probabilidade_de_ser_ruim))
print(p)
```
