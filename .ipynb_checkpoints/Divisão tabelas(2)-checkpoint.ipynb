```python
def cleanup(text):
    punctuation = '[!\ \n-.:?;,@]' # Note que os sinais [] são delimitadores de um conjunto.
    pattern = re.compile(punctuation)
    novoindex = []
    for frase in text:
         novoindex.append(re.sub(pattern, ' ', frase))
    return novoindex   
```

```python
p_bom=0
p_ruim=0
p_irrelevante=0
for i in palavras:
    if i not in tabela_freq2a: 
        prob_bom[i]=0
        tabela_freq2a[i]=1e-8
    if i not in tabela_freq3a: 
        prob_ruim[i]=0
        tabela_freq3a[i]=1e-8
    if i not in tabela_freq4: 
        prob_irrelevante[i]=0
        tabela_freq4a[i]=1e-8
    p_bom+=(np.log(prob_bom[i]/total_irrelevantes+(1/total_valores)))*probabilidade_de_ser_boa
    p_ruim+=(np.log(prob_ruim[i]/total_bom+(1/total_valores)))*probabilidade_de_ser_ruim
    p_irrelevante+=(np.log(prob_irrelevante[i]/total_bom+(1/total_valores)))*probabilidade_de_ser_irrelevante
    
print(p_bom)
print(p_ruim)
print(p_irrelevante)
```

```python
def concatena(tweet):
    p_bom = 0
    p_ruim = 0
    p_irrelevante = 0
    for pa in tweet:
        if pa not in prob_bom: 
            prob_bom[pa] = 1e-8
        if pa not in prob_ruim: 
            prob_ruim[pa] = 1e-8
        if pa not in prob_irrelevante: 
            prob_irrelevante[pa] = 1e-8
        p_bom += np.log(prob_bom[pa])
        p_ruim += np.log(prob_ruim[pa])
        p_irrelevante += np.log(prob_irrelevante[pa])
    p_bom += np.log(probabilidade_de_ser_boa)
    p_ruim += np.log(probabilidade_de_ser_ruim)
    p_irrelevante += np.log(probabilidade_de_ser_irrelevante) 
    
    if p_bom > p_ruim and p_bom > p_irrelevante:
        return 1
    elif p_ruim > p_bom and p_ruim > p_irrelevante:
        return 2
    else:
        return 0
```
