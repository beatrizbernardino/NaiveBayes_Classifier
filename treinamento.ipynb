```python
planilha1=pd.read_excel("era uma vez em hollywood.xlsx", sheet_name='Treinamento')
```

```python
excel1=excelcomrt.loc[excelcomrt['Classificação']!=3]
```

```python
planilha2=pd.read_excel("era uma vez em hollywood2.xlsx", sheet_name='Treinamento')
```

```python
excel2=planilha2.set_index('Treinamento')
```

```python
tabela = excel1.append(excel2)
```

```python
novoindex=[]
def cleanup(text):
    punctuation = '[!\ \n-.:?;,@"  "]' # Note que os sinais [] são delimitadores de um conjunto.
    pattern = re.compile(punctuation)
    for frase in text:
         novoindex.append(re.sub(pattern, ' ', frase))
    return novoindex
   
```

```python
novoindex2=cleanup(tabela.index)
```

```python
tabela["tweets"]=novoindex2
tabela.set_index("tweets")
```
